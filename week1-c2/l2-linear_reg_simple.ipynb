```python
# x - m2 of the flat
# y - price of the flat in Mrub
x = [53.6, 81.1, 98.3, 56.0, 53.9, 54.7, 69, 39.1, 63.2, 30.5, 79, 71]
y = [20.9, 13.5, 34.8, 12.8, 11.45, 24.0724, 25.99, 9.5, 13.3, 8.4, 15.3, 16.6]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x, y)
plt.show()
```

```python
def linear_regression(x, w1):
  predictions = []
  for x1 in x:
    predictions.append(w1 * x1)
  return predictions
```

```python
predictions_050 = linear_regression(x, 0.5)
predictions_025 = linear_regression(x, 0.25)
```

```python
plt.scatter(x, y)
plt.plot(x, predictions_050, color='r') # y^ - predictions
plt.plot(x, predictions_025, color='g') # y^ - predictions
plt.show()
```

```python
def loss_mse(y, predictions):
  l = 0
  for yi, pi in zip(y, predictions):
    l += (yi - pi)**2
  l = l / len(y)
  return l
```

```python
print("MSE loss for w1 = 0.50: ", loss_mse(y, predictions_050))
print("MSE loss for w1 = 0.25: ", loss_mse(y, predictions_025))
```

```python
import numpy as np
w1_list = [0.05, 0.10, 0.15, 0.20, 0.25, 0.30, 0.35, 0.40, 0.45, 0.5]
w1 = np.linspace(0.05, 0.3, 10)
w2 = np.linspace(0.305, 0.5, 20)
w1_list = np.hstack((w1, w2))
ll = []
for w1 in w1_list:
  preds = linear_regression(x, w1)
  l = loss_mse(y, preds)
  print("MSE loss for w1 = ", w1, " : ", l)
  ll.append(l)
```

```python
plt.scatter(w1_list, ll)
plt.show()
```

```python
def linear_regression_gradient(y, x, predictions):
  grad = 0
  for yi, xi, p in zip(y, x, predictions):
    grad += (2 * (yi - p) * (-xi))
  grad = grad / len(y)
  return grad
```

```python
# gradient descend
alpha = 0.0001
w = 0.5
nstep = 10
ll = []
for i in range(nstep):
  predictions = linear_regression(x, w)
  grad = linear_regression_gradient(y, x, predictions)
  w = w - alpha * grad
  l = loss_mse(y, predictions)
  ll.append(l)
  print(w, l)
```
